Exemplo 2 – Usando o retorno em outra função

```python
def somatresnotas():
  nota1 = float(input("Digite sua primeira nota: "))
  nota2 = float(input("Digite sua segunda nota: "))
  nota3 = float(input("Digite sua terceira nota: "))
  total = nota1 + nota2 + nota3
  return total
```

```python
def mediatresnotas():
  soma = somatresnotas()
  media = soma/3
  return media
```

```python
print("A média das notas seria: ", mediatresnotas())
```

Exercicio 4:

```python
def calcular_potencias(n):
  quadrado = n * n
  cubo = n * n * n
  print(f"O quadrado de {n} é: ", quadrado)
  print(f"O cubo de {n} é: ", cubo)
  return quadrado, cubo
```

```python
print(calcular_potencias(3))
```

Exercicio 5:

```python
def contagem_regressiva(numero_de_inicio):
  num_inicio = numero_de_inicio
  print(num_inicio)
  while num_inicio != 0:
    num_inicio = num_inicio - 1
    print(num_inicio)

contagem_regressiva(10)
```
